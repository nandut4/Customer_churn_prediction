```python
import pandas as pd
```

```python
data = pd.read_csv(r'C:\Users\Gouthami\Documents\DSML\Customer_churn_prediction\data\WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
data.head()
```

```python
data.drop('customerID', axis=1, inplace=True)
```

```python
data['TotalCharges'] = data['TotalCharges'].str.strip()
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
```

```python
data.dropna(axis=0, inplace=True)
```

```python
data.nunique()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in data.columns:
    if(data[col].dtype == 'str'):
        if(data[col].nunique() == 2):
            data[col] = le.fit_transform(data[col])
```

```python
data.head()
```

```python

# columns with more than 2 categories
multi_cat_cols = [
    col for col in data.columns
    if data[col].dtype == 'str' and data[col].nunique() > 2
]

# apply one hot encoding
data = pd.get_dummies(data, columns=multi_cat_cols, drop_first=True)
```
